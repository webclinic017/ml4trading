```python
def is_trading_day_and_time_today():
    nyse       = mcal.get_calendar('NYSE')
    early      = nyse.schedule(start_date='2021-01-01', end_date='2021-12-31')
    valid_days = nyse.valid_days(start_date='2021-01-01', end_date='2021-12-31')
    now        = datetime.now(tz=pytz.timezone('US/Eastern'))
    td         = date.today()
    valid_day = (td in valid_days)
    valid_time = nyse.open_at_time(early, now)
   
    valid_day_and_time = (valid_day and valid_time)
    
    # mm/dd/y
    td_string = td.strftime("%Y-%m-%d %H:%M:%S %z %Z")
    dt_string = now.strftime("%Y-%m-%d %H:%M:%S %z %Z")
    print("here_date={}\nmkt_now={}\nvalid_day={} valid_time={}\nis_trading_day_and_time={}".format(td_string, dt_string, valid_day, valid_time, (valid_day and valid_time)))

    return valid_day_and_time
```
